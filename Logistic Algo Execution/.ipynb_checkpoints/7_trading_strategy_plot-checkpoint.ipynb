```python
#tradeplotttt
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt

file=pd.read_csv("/Users/chen/Desktop/homeworks/aapl_signal.csv")

def trade_e(file,cost:float,threshold:float):

    N=0
    wealth=1000
    W=[]

    Price=file['average_price']
    signal=file['signal_logistic']
    strength=file['logistic_prob']

    sig=signal*(strength>threshold)
    position=0*Price

    for i in range(len(position)):

        if wealth<=0:
            W.append(wealth)
            continue
        if i==0:
            position[i]=sig[i]*wealth
            W.append(wealth)
            N+=abs(sig[i])
            continue
        else:
            position[i]=position[i-1]
        if sig[i]==1  and position[i-1]==0:
            wealth=wealth*(1-cost)
            N+=1
            position[i]=wealth
        elif sig[i]==-1 and position[i-1]==0:
            wealth=wealth*(1-cost)
            N += 1
            position[i]=-wealth
        elif sig[i]==0  and position[i-1]!=0:
            wealth=wealth*(1-cost)
            N += 1
            position[i]=0
        elif sig[i]*position[i-1]<0:
            wealth=wealth*(1-cost)**2
            N += 2
            position[i]=wealth*sig[i]
            # position[i]=0

        wealth+=position[i-1]*(Price[i]-Price[i-1])/Price[i-1]
        # print(wealth)
        W.append(wealth)

    print(N)
    plt.figure(dpi=128,figsize=(5,3.09))
    plt.subplot(111)

#     plt.scatter(file.index,W,c=W,cmap=plt.cm.Blues,s=0.5)
    plt.plot(W,c='b',alpha=0.75,label="Wealth")
    plt.xlabel("Min")
    plt.ylabel("Wealth")
    plt.title("Return: " + str(wealth / 10 - 100) + "% cost/threshold: " + str(cost) + "/" + str(threshold))

#     plt.subplot(212)
    plt.twinx()
#     plt.scatter(file.index,Price,c=Price,cmap=plt.cm.Reds,s=0.5)
    plt.plot(Price,c='b',alpha=0.25,label="Price")
#     plt.xlabel("Min")
    plt.ylabel("Price")
#     plt.title("close price")

    

    plt.show()
def trade_ee(file,cost:float,threshold:float):

    N=0
    wealth=1000
    W=[]

    Price=file['average_price']
    signal=file['signal_logistic']
    strength=file['logistic_prob']

    sig=signal*(strength>threshold)
    position=0*Price

    for i in range(len(position)):

        if wealth<=0:
            W.append(wealth)
            continue
        if i==0:
            position[i]=sig[i]*wealth
            W.append(wealth)
            N+=abs(sig[i])
            continue
        else:
            position[i]=position[i-1]
        if sig[i]==1  and position[i-1]==0:
            wealth=wealth*(1-cost)
            N+=1
            position[i]=wealth
        elif sig[i]==-1 and position[i-1]==0:
            wealth=wealth*(1-cost)
            N += 1
            position[i]=-wealth
        elif sig[i]==0  and position[i-1]!=0:
            wealth=wealth*(1-cost)
            N += 1
            position[i]=0
        elif sig[i]*position[i-1]<0:
            wealth=wealth*(1-cost)**2
            N += 2
            position[i]=wealth*sig[i]
            # position[i]=0

        wealth+=position[i-1]*(Price[i]-Price[i-1])/Price[i-1]
        # print(wealth)
        W.append(wealth)

    print(N)
    plt.figure(dpi=128,figsize=(5,3.09))
    plt.subplot(211)

    plt.scatter(file.index,W,c=W,cmap=plt.cm.Blues,s=0.5)
#     plt.plot(W,c='b',alpha=0.75,label="Wealth")
    plt.xlabel("Min")
    plt.ylabel("Wealth")
    plt.title("Return: " + str(wealth / 10 - 100) + "% cost/threshold: " + str(cost) + "/" + str(threshold))

    plt.subplot(212)
#     plt.twinx()
    plt.scatter(file.index,Price,c=Price,cmap=plt.cm.Reds,s=0.5)
#     plt.plot(Price,c='b',alpha=0.25,label="Price")
#     plt.xlabel("Min")
    plt.ylabel("Price")
#     plt.title("close price")

    

    plt.show()

```

```python
def my_plot(df,ticker,cost=0):
    Average_P=df['average_price']
    W=df['wealth']
    plt.figure(dpi=128,figsize=(5,3.09))
    plt.subplot(211)

#     plt.scatter(df.index,W,c=W,cmap=plt.cm.Blues,s=0.5)
    plt.plot(W,c='navy',alpha=0.75,label="Wealth")
    plt.xlabel("Min")
    plt.ylabel("Wealth")
    if cost:
        plt.title(ticker+" Return: " + str(round(list(W)[-1] / 10 - 100,2))+"% (with cost)" )
    else:
        plt.title(ticker+" Return: " + str(round(list(W)[-1] / 10 - 100,2))+"% (without cost)" )

    plt.subplot(212)
#     plt.twinx()
#     plt.scatter(df.index,Average_P,c=Average_P,cmap=plt.cm.Reds,s=0.5)
    plt.plot(Average_P,c='b',alpha=0.25,label="Price")
#     plt.xlabel("Min")
    plt.ylabel("Average Price")
#     plt.title("close price")

    
#     plt.show()
    
```
